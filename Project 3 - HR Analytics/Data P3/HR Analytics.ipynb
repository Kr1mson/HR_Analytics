```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\parth\OneDrive\Desktop\Project 3 - HR Analytics\Data P3\HR-Employee-Attrition.csv")
hr_list=df.copy()
hr_list.head()
redundant=['Over18','EmployeeCount','RelationshipSatisfaction','WorkLifeBalance','StandardHours']
hr_list.drop(columns=redundant,axis=1,inplace=True)
hr_list.drop_duplicates()
hr_list.isna().sum()
hr_list.dropna()
hr_list.describe()
#hr_list.columns
#hr_list.to_csv(r"C:\Users\parth\OneDrive\Desktop\Project 3 - HR Analytics\Data P3\HR-Employee-Attrition1.csv")
```

```python
corr_matrix=hr_list.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix,cmap='bwr', annot=True, fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
response_counts = hr_list['OverTime'].value_counts()
plt.bar(response_counts.index, response_counts.values, color=['red','green'])
plt.title('Overtime')
plt.xlabel('Response')
plt.ylabel('Employee Count')
plt.show()
```

```python
response_counts = hr_list['MaritalStatus'].value_counts()
plt.bar(response_counts.index, response_counts.values, color=['orange','grey','blue'])
plt.title('Marital Status')
plt.ylabel('Employee Count')
plt.show()
```

```python

response_counts = hr_list['JobRole'].value_counts()
plt.bar(response_counts.index, response_counts.values,)
plt.title('Job Role')
plt.ylabel('Employee Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
response_counts = hr_list['BusinessTravel'].value_counts()
plt.bar(response_counts.index, response_counts.values,color=['darkgrey','yellow','red'])
plt.title('Business Travel')
plt.ylabel('Employee Count')
plt.show()
```

```python
age_bins = np.arange(18, 60,5)
age_labels = [f"{start}-{start+4}" for start in age_bins[:-1]]
hr_list['Age_Group'] = pd.cut(hr_list['Age'], bins=age_bins, labels=age_labels, right=False)
grouped_data = hr_list.groupby(['Age_Group', 'OverTime']).size().unstack(fill_value=0)
grouped_data.plot(kind='bar', stacked=True)
plt.title('Overtime Preferences by Age Group (10-year intervals)')
plt.xlabel('Age Group')
plt.ylabel('Employee Count')
plt.show()
```
